# Data Loading

```python
df_res_duplicate = pd.read_csv('../Experiment_data/20_duplicate/Qwen2.5-3B-Instruct/results_VANILLA.csv')

all_res = {}

init_res = {'Exact': 0,
 'ContextCite32': 0,
 'KernelShap32': 0,
 'BetaShap (U)32': 0,
 'TMC32': 0,
 'ContextCite64': 0,
 'KernelShap64': 0,
 'BetaShap (U)64': 0,
 'TMC64': 0,
 'ContextCite100': 0,
 'KernelShap100': 0,
 'BetaShap (U)100': 0,
 'TMC100': 0,
 'LOO': 0}

available_models = {"qwen_3B" : "Qwen/Qwen2.5-3B-Instruct", 
                    # "llama_3B" : "meta-llama/Llama-3.2-3B-Instruct", 
                    "llama_8B" : "meta-llama/Llama-3.1-8B-Instruct", 
                    "mistral_7B" : "mistralai/Mistral-7B-Instruct-v0.3"}


for data_type in ["synergy", "duplicate", "complementary"]: 
    for model in available_models.keys(): 
        df = pd.read_csv(f'../Experiment_data/20_{data_type}/{available_models[model].split("/")[1]}/results_VANILLA.csv')
        res = init_res.copy()
        for i in df.index: 
            for method, score in eval(df.precision_top_k.loc[i])[0].items(): 
                res[method[1]] += score
        all_res[f"{data_type}_{available_models[model].split('/')[0]}"] = {k : v/20 for k,v in res.items()}
```

```python
def create_id(df : pd.DataFrame) -> pd.DataFrame: 
    letters = list(string.ascii_uppercase[: len(eval(df.context.loc[0]))])
    df["doc_id"] = np.vstack([letters]*df.shape[0]).tolist()
    return df
```

```python
df_bio = pd.read_csv('../data/BIOASK.csv')
# df_bio.drop(df_bio.columns[0], axis = 1, inplace=True)
df_bio = create_id(df_bio.copy())
df_bio.to_csv('../data/BIOASK.csv', index = False)
```

```python
df_nq = pd.read_csv("../data/NQ.csv")
df_nq = create_id(df_nq.copy())
df_nq.to_csv('../data/NQ.csv', index = False)
```
