```python
import pandas as pd
import ast
import copy
import json
```

```python
synergy = pd.read_csv(
    r"C:\Users\gabri\Desktop\code\tesi\LLMX\data\synthetic_data\gabriele_S_20.csv"
)

complementary = pd.read_csv(
    r"C:\Users\gabri\Desktop\code\tesi\LLMX\data\synthetic_data\complementary.csv"
)

duplicate = pd.read_csv(
    r"C:\Users\gabri\Desktop\code\tesi\LLMX\data\synthetic_data\duplicate.csv"
)


print((eval(synergy.context[0])), type(complementary.context), type(duplicate.context))

synergy = synergy.sample(n=3, random_state=42)

complementary = complementary.sample(n=3, random_state=42)

duplicate = duplicate.sample(n=3, random_state=42)

duplicate["id"] = [
    ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"] for _ in range(len(duplicate))
]

complementary.drop("id", axis=1)

complementary["id"] = [
    ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
    for _ in range(len(complementary))
]

synergy.drop("id", axis=1)

synergy["id"] = [
    ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"] for _ in range(len(synergy))
]


df = pd.concat([duplicate, complementary, synergy])

print(df.info())
```

```python
augmented_rows = []

for index, row in df.iterrows():
    original_question = row["question"]
    original_context_str = row["context"]
    original_answer = row["answer"]
    original_id = row["id"]  # The 'id' column refers to sentence IDs, keep it as is.
    context_sentences = ast.literal_eval(original_context_str)
    # Store the original row first
    augmented_rows.append(
        {
            "question": original_question,
            "context": original_context_str,  # Keep original string format
            "answer": original_answer,
            "id": original_id,
        }
    )

    # Variation 1: Flip the first two sentence
    shuffled_v1 = context_sentences.copy()
    shuffled_id_v1 = copy.deepcopy(original_id)
    shuffled_v1[0], shuffled_v1[1] = shuffled_v1[1], shuffled_v1[0]
    shuffled_id_v1[0], shuffled_id_v1[1] = shuffled_id_v1[1], shuffled_id_v1[0]
    augmented_rows.append(
        {
            "question": original_question,
            "context": json.dumps(shuffled_v1),  # Convert list back to JSON-like string
            "answer": original_answer,
            "id": shuffled_id_v1,
        }
    )

    # Variation 2: Switch the second sentence with the last

    shuffled_v2 = context_sentences.copy()
    shuffled_id_v2 = copy.deepcopy(original_id)

    shuffled_v2[1], shuffled_v2[-1] = shuffled_v2[-1], shuffled_v2[1]
    shuffled_id_v2[1], shuffled_id_v2[-1] = shuffled_id_v2[-1], shuffled_id_v2[1]
    augmented_rows.append(
        {
            "question": original_question,
            "context": json.dumps(shuffled_v2),
            "answer": original_answer,
            "id": shuffled_id_v2,
        }
    )


# 4. Create a new DataFrame from the augmented rows
result_df = pd.DataFrame(augmented_rows)

print("--- Augmented DataFrame (First few rows and total count) ---")
print(f"Total rows generated: {len(result_df)}")
if len(result_df) > 0:
    # Display one full example (original + 3 variants)
    if len(result_df) >= 4:
        print("\nExample of one original row and its 3 variants:")
        print(result_df.head(4).to_string())

    # To see all 12 rows if you sampled 3 and all shuffles were possible:
    print("\n--- Full Augmented DataFrame ---")
    print(result_df.to_string())  # .to_string() helps print all rows/cols

# If you want to save this to a new CSV:
# result_df.to_csv('augmented_data.csv', index=False)
```
